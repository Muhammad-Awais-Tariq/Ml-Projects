```python
import numpy as np
import pandas as pd
```

```python
titanic_test_df = pd.read_csv("F://Ml-Projects//Data//Titanic//test.csv")
df = titanic_test_df.copy()
```

```python
df.head()
```

```python
df.drop(columns=["PassengerId" , "Name" , "Ticket" ] , inplace=True)
```

```python
df.head()
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
)
```

```python
df["age"] = df["age"].fillna(df["age"].median())
```

```python
df["has_cabin"] = df["cabin"].notna().astype(int)
df.drop(columns=["cabin"] , inplace=True)
df
```

```python
df["embarked"] = df["embarked"].fillna(df["embarked"].mode()[0])
```

```python
df["fare"] = df["fare"].fillna(df["fare"].median())
```

```python
df["sex"] = df["sex"].map({"female" : 0 , "male" : 1})
```

```python
df = pd.get_dummies(data=df , columns=["embarked"] , drop_first=True) 
```

```python
df[["embarked_Q","embarked_S"]] = df[["embarked_Q","embarked_S"]].astype(int)
```

```python
df
```

```python
df.to_csv("F://Ml-Projects//Data//Titanic//cleaned_test_data.csv")
```
