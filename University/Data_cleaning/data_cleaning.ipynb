```python
import numpy as np
import pandas as pd
import sweetviz as sv
```

```python
student_record = pd.read_csv("F://Ml-Projects//Data//University//students_record_raw.csv")
report = sv.analyze(student_record)
report.show_html("F://Ml-Projects//University//Data_cleaning//Reports//raw_data_report.html" , open_browser=False)
```

```python
df = student_record.copy()
df
```

```python
df.drop(columns=["student_id" , "student_email"] , inplace=True)
```

```python
df["attendance_pct"] = df["attendance_pct"].fillna(df["attendance_pct"].median())
df["lab_score"] = df["lab_score"].fillna(df["lab_score"].median())
df["study_hours_per_week"] = df["study_hours_per_week"].fillna(df["study_hours_per_week"].median())
```

```python
df.isnull().sum()
```

```python
df = df[df["mid_score"] <= 50]
```

```python
df
```

```python
final_report = sv.analyze(df)
final_report.show_html("F://Ml-Projects//University//Data_cleaning//Reports//cleaned_data_report.html",open_browser=False)
```

```python
df.to_csv("F://Ml-Projects//Data//University//cleaned_students_record.csv")
```
